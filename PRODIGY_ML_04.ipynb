```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
```

```python
np.random.seed(42)
dates = pd.date_range(start="2023-01-01", periods=1000, freq='D')
tweets = ["I love this brand!", "This product is terrible.", "Had a great experience with this service.", "I am not happy with the quality.", "Best purchase ever!", "Worst customer service."]
```

```python
data = pd.DataFrame({
    'date': np.random.choice(dates, size=1000),
    'tweet': np.random.choice(tweets, size=1000)
})
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity
```

```python
data['sentiment'] = data['tweet'].apply(get_sentiment)
```

```python
daily_sentiment = data.groupby('date')['sentiment'].mean().reset_index()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=daily_sentiment, x='date', y='sentiment')
plt.title('Daily Sentiment Analysis')
plt.xlabel('Date')
plt.ylabel('Average Sentiment')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['sentiment'], bins=20, kde=True)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
