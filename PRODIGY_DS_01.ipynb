```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
ages = np.random.normal(loc=35, scale=10, size=1000).astype(int)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ages, bins=20, kde=True, color='blue')
```

```python
plt.title('Distribution of Ages in Population', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
```

```python
plt.show()
```

```python
genders = ['Male', 'Female']
counts = [480, 520]
```

```python
plt.figure(figsize=(8, 6))
plt.bar(genders, counts, color=['blue', 'pink'])
```
